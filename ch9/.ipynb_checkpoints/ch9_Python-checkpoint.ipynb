### Data analysis script for NetLogo output

A set of visualisations for a standard table output of NetLogo: scatterplot, barplot, and lineplot.

This is an example script used in chapter 9 of Romanowska, I., Wren, C., Crabtree, S. 2021 Agent-Based Modeling for Archaeology: Simulating the Complexity of Societies. Santa Fe Institute Press.

Code blocks: 9.7, 9.9, 9.11, 9.13

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("deep")
sns.set()

data = pd.read_csv('Gaul_Model experiment-table.csv', skiprows = 6)
```

```python
# Check how everything looks like
data.head()
```

### SCATTERPLOT

```python
sns.scatterplot(x = "weighted-trade-choice", y = "count turtles", data = data)
```

### BARPLOT

```python
plt.figure(figsize = (10,7))
sns.barplot( x = "weighted-trade-choice", y = 'count turtles', hue = 'metals?',data = data)
```

```python
# read file, the first rows do not contain data
curve_data = pd.read_csv("metals_ true_repro3_13.csv", skiprows=17)
```

```python
# meaningful column names
curve_data.rename(columns ={"y":"Etruscan", "y.1":"Greek" }, inplace = True)
```

```python
curve_data.head()
```

### LINEPLOT Basic version

```python
plt.figure(figsize=(15, 10))
sns.lineplot( y = "Etruscan", x = "x", data = curve_data)
sns.lineplot( y = "Greek", x = "x.1", data = curve_data)
```

### LINEPLOT Better version

```python
plt.figure(figsize=(15, 10))
sns.lineplot( y = "Etruscan", x = "x", data = curve_data, label = "Etruscan Wine")
sns.lineplot( y = "Greek", x = "x.1", data = curve_data, label = "Greek Wine")

plt.legend()
plt.xlabel('Year since arrival')
plt.ylabel('Wine')
# save figure 
plt.savefig("ch9_the_curve.png", dpi = 300)
```
